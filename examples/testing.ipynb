```python
from phylogenetics import TreeProject
```

```python
p = TreeProject('project')
```

```python
p.read_data(path='PF08793_seed.fasta', schema='fasta', dtype='tips')
```

```python
p.run_tree()
```

```python
p.run_reconstruction()
```

```python
p.save()
```

```python
p.draw_tree(width=700,
    tip_labels='id',
    tip_labels_align=True,
    use_edge_lengths=True,
    node_labels='id')
```
